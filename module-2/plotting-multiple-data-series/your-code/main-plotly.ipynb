# Plotting Multiple Data Series

Complete the following set of exercises to solidify your knowledge of plotting multiple data series with pandas, matplotlib, and seaborn. Part of the challenge that comes with plotting multiple data series is transforming the data into the form needed to visualize it like you want. For some of the exercises in this lab, you will need to transform the data into the form most appropriate for generating the visualization and then create the plot.

```python
import cufflinks as cf
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df = pd.read_csv('../data/liquor_store_sales.csv')
display(df.head())
df.info(memory_usage='deep')
```

```python
df.nunique()
```

## 1. Create a bar chart with bars for total Retail Sales, Retail Transfers, and Warehouse Sales by Item Type.

```python
cf.go_offline()
```

```python

df_bar=df[['RetailSales','RetailTransfers','WarehouseSales','ItemType']].groupby('ItemType',as_index=False).sum()

px.bar(df_bar, x="ItemType", y=['RetailSales','RetailTransfers','WarehouseSales'], barmode="group")
```

## 2. Create a horizontal bar chart showing sales mix for the top 10 suppliers with the most total sales. 

```python
df[['RetailSales','RetailTransfers','WarehouseSales']].sum().sum()
```

```python

df['Total Sales']=df['RetailSales'] + df['RetailTransfers'] + df['WarehouseSales']

df_salesmix_top10=df[['RetailSales','RetailTransfers','WarehouseSales','Total Sales','Supplier']].groupby('Supplier',as_index=False)\
                        .sum().sort_values(by='Total Sales',ascending=False).iloc[:10]


df_salesmix_top10.set_index('Supplier').plot(kind='barh', fontsize=12, figsize=(12,12))
plt.show()
```

## 3. Create a multi-line chart that shows average Retail Sales, Retail Transfers, and Warehouse Sales per month over time.

```python
df['Month'] = df['Month'].astype('str')
df['Year'] = df['Year'].astype('str')


def month(x):
    if len(x)==2:
        return x
    else:
        return f'0{x}'

df['Month_Year']=df['Month'].apply(month)
df['Month_Year']=  df['Year'] + '/'+  df['Month_Year']

df_month_year=df[['RetailSales','RetailTransfers','WarehouseSales','Month_Year']].groupby('Month_Year',as_index=False).mean()

df_month_year.set_index('Month_Year').plot(kind='line', fontsize=12, figsize=(14,8),xticks=df_month_year.index)
plt.title(label='Average by Month')
plt.show()

```

## 4. Plot the same information as above but as a bar chart.

```python
df_month_year.set_index('Month_Year').plot(kind='bar', fontsize=12, figsize=(14,8),title='Average by Month')
plt.xticks(df_month_year.index, df_month_year['Month_Year'], rotation=0)
plt.show()
```

## 5. Create a multi-line chart that shows Retail Sales summed by Item Type over time (Year & Month).

*Hint: There should be a line representing each Item Type.*

```python
df_month_year_retailsales=df[['RetailSales','ItemType','Month_Year']].groupby(['ItemType','Month_Year'],as_index=False).sum()

fig = plt.figure(figsize=(14, 8))
sns.lineplot(data=df_month_year_retailsales, x='Month_Year', y='RetailSales', hue='ItemType')
plt.title(label='Retail Sales summed by Item Type')
plt.show()
```

## 6. Plot the same information as above but as a bar chart.

```python
fig = plt.figure(figsize=(14, 8))
sns.barplot(data=df_month_year_retailsales, x='Month_Year', y='RetailSales', hue='ItemType')
plt.title(label='Retail Sales summed by Item Type')
plt.show()
```

## 7. Create a scatter plot showing the relationship between Retail Sales (x-axis) and Retail Transfers (y-axis) with the plot points color-coded according to their Item Type.

*Hint: Seaborn's lmplot is the easiest way to generate the scatter plot.*

```python
df_sales_transfers=df[['RetailSales','RetailTransfers','ItemType']]

sns.lmplot(data=df_sales_transfers, x='RetailSales', y='RetailTransfers', hue='ItemType',height=8,palette='bright')
plt.title(label='Relationship between Retail Sales (x-axis) and Retail Transfers (y-axis)')
plt.show()


```

## 8. Create a scatter matrix using all the numeric fields in the data set with the plot points color-coded by Item Type.

*Hint: Seaborn's pairplot may be your best option here.*

```python
df_numeric=df.select_dtypes('number')
df_numeric['ItemType']=df['ItemType']


sns.pairplot(df_numeric,height=3,hue='ItemType', palette='bright')

plt.show()

```
